## Preprocessing for prediction

```
#dropping all the rows with any null value 
train_df= train_df.dropna(axis=0,how='any' )
```

```
train_df.info()
```

```
train_df.head(5)
```

#Prediction

```
df = train_df
```

```
#Dependicies for prediction
from sklearn import preprocessing
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.feature_column import numeric_column
from tensorflow.estimator import LinearClassifier
from tensorflow.data import Dataset
from sklearn import preprocessing
```

```
#changing Parch and Sex in Numeric form 
le = preprocessing.LabelEncoder()
le.fit(df['Sex'])
#le.fit(df['Parch'])
df['Sex']= le.transform(df['Sex'])
#df['Parch'] =le.transorm(df['Parch'])
```

```
#splitting the training data into train and validation
train , val = train_test_split(df,test_size=0.2,
                               stratify=df['Survived'],
                               random_state=42)
```

```
#changing the columns to fit tensorflow
features= ['Pclass','Sex','Age', 'SibSp']
targets = ['Survived']


new_train  = df[features]
new_train['Survived'] = df[targets]


features_col=[]
for feature_name in new_train.columns[:-2]:
  features_col.append(numeric_column(str(feature_name)))
features_col
```

```
classifier = LinearClassifier(feature_columns=features_col, n_classes=2)
```

```
features
```

```
train.info()
```

```
features_dict = {}
for i in features:
  features_dict[str(i)] = train[i] 
```

```
#Training the Classifer
def training_input():
  features_dict = {}
  for i in features:
    features_dict[str(i)] = train[i]
  labels = train['Survived']
  training_ds = Dataset.from_tensor_slices((features_dict, labels))
  training_ds= training_ds.shuffle(buffer_size=100)
  training_ds=training_ds.batch(100)
  training_ds=training_ds.repeat(5)
  return training_ds
classifier.train(training_input)
```

```
from tensorflow.data import Dataset

#training the classifier 
def val_input():
  features_dict = {}
  for i in features:
    features_dict[str(i)] = val[i]
 
  labels = val['Survived']
  
  training_ds = Dataset.from_tensor_slices((features_dict, labels))
  training_ds = training_ds.batch(100)
  training_ds = training_ds.repeat(1)

  return training_ds

classifier.evaluate(val_input)
```

```
predictions_iterator_one = list(classifier.predict(val_input))
predictions_iterator_one
```

```
predictions = [p['logistic'][0] for p in predictions_iterator_one]
print(predictions)
```

```
threshold = .60
survived=[]
for i in predictions:
  if (i > threshold):
    survived.append(1)
  else:
    survived.append(0)
    
```

```
from sklearn.metrics import recall_score
recall_score(val['Survived'], survived)
```

```
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score

precision_score(val['Survived'], survived)
```

```
accuracy_score(val['Survived'], survived)
```

```
from sklearn.metrics import average_precision_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
from inspect import signature
```

```
# building precision vs. recall chart
average_precision = average_precision_score(val['Survived'], predictions, average=None) ###test_df['Survived'])
precision, recall, thresholds = precision_recall_curve(val['Survived'], predictions)

# print('Average precision-recall score: {0:0.2f}'.format(
#       average_precision))


# In matplotlib < 1.5, plt.fill_between does not have a 'step' argument
step_kwargs = ({'step': 'post'}
               if 'step' in signature(plt.fill_between).parameters
               else {})
plt.step(recall, precision, color='b', alpha=0.2,
         where='post')
plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(
          average_precision))
plt.show()
```
